```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# find total number of schools
total_schools = len(school_data["school_name"])
```

```python
# find total number of students
total_students = len(student_data["student_name"])
```

```python
# total budget
budget = sum(school_data["budget"])
```

```python
#average math score
average_math = sum(student_data["math_score"])/len(student_data["math_score"])
```

```python
#average reading score
average_reading = sum(student_data["reading_score"])/len(student_data["reading_score"])
```

```python
# number of students passing math
passing_math = sum(student_data.math_score>=70)
```

```python
#percentage of students passing math
percent_math = passing_math / len(student_data.math_score) *100
```

```python
#number of students passing reading
passing_reading = sum(student_data.reading_score>=70)
```

```python
#percentage of students passing reading
percent_reading = passing_reading / len(student_data.reading_score) *100
```

```python
# Overall percent of students passing both math and reading
playing_df = student_data.loc[(student_data["reading_score"] >=70) & (student_data["math_score"] >= 70),:]

overall_passing = len(playing_df) / len(student_data) *100
```

```python
#create dataframe to display all the above findings
percentages_df = pd.DataFrame(
    {"Total Schools" : [total_schools], "Total Students": [total_students],"Total Budget": [budget],
     "Average Math Score": [average_math], "Average Reading": [average_reading],
    "% Passing Math": [percent_math], "% Passing Reading": [percent_reading], "% Overall Passing": [ overall_passing]})
percentages_df
```

```python
#format for comma separation and dollar sign
percentages_df["Total Students"] = percentages_df["Total Students"].map("{:,}".format)
percentages_df["Total Budget"] = percentages_df["Total Budget"].map("${:,.2f}".format)
percentages_df
```
